```python
import pandas as pd

def f(expr,x,y):
    return eval(expr,{'x':x,'y':y})

def euler(expr,xi,yi,xf,h):
    # step_size=list(map(int,input('Step size:').split()))
    n=int((xf-xi)/h)
    x=xi;y=yi
    # print('%f \t %f'%(x,y))
    data=[]
    data.append([round(x,5),round(y,5)])
    for i in range(1,n+1):
        y=y+f(expr,x,y)*h
        x=x+h
        # print('%f \t %f'%(x,y))
        data.append([round(x,5),round(y,5)])
    df=pd.DataFrame(data,columns=['X','Y'])
    display(df)
    print('Root is %.6f'%(y))

expr='-2.2067e-12*(y**4-81e8)'
xi=0
yi=1200
xf=480
h=120

euler(expr,xi,yi,xf,h)
```
