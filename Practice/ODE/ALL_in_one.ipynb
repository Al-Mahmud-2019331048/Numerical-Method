```python
import matplotlib.pyplot as plt


def f(expr,x,y):
    val=eval(expr)
    return val


x_plot=[];y_euler=[];y_heun=[];y_midpoint=[];y_ralston=[]


def Euler(x,y,n):
    x_plot.append(x);y_euler.append(y)
    print("Euler Method:")
    for i in range(1,n+1):
        y=y+f(expr,x,y)*h
        x=x+h
        print('%f \t %f'%(x,y))
        x_plot.append(x);y_euler.append(y)

def Heun(x,y,n):
    y_heun.append(y)
    print("Heun's Method:")
    for i in range(1,n+1):
        k1=f(expr,x,y)
        k2=f(expr,x+h,y+k1*h)
        y=y+(k1/2+k2/2)*h
        print('%f \t %f'%(x,y))
        y_heun.append(y)

def Midpoint(x,y,n):
    y_midpoint.append(y)
    print("Midpint Method:")
    for i in range(1,n+1):
        k1=f(expr,x,y)
        k2=f(expr,x+h/2,y+k1*h/2)
        y=y+k2*h
        print('%f \t %f'%(x,y))
        y_midpoint.append(y)


def Ralston(x,y,n):
    y_ralston.append(y)
    print("Ralston's Method:")
    for i in range(1,n+1):
        k1=f(expr,x,y)
        k2=f(expr,x+3*h/4,y+3*k1*h/4)
        y=y+(k1/3+2*k2/3)*h
        print('%f \t %f'%(x,y))
        y_ralston.append(y)


xi=0
yi=1200
xf=480
h=120
n=int((xf-xi)/h)
expr='-2.2067*10**(-12)*(y**4-81*10**8)'


x=xi;y=yi

Euler(x,y,n)
Heun(x,y,n)
Midpoint(x,y,n)
Ralston(x,y,n)

plt.plot(x_plot,y_euler,"",x_plot,y_heun,"",x_plot,y_midpoint,"",x_plot,y_ralston)
plt.legend(['Euler',"Heun's","Midpoint","Ralston's"])
plt.xlabel(x)
plt.ylabel(y)
plt.show()
```
