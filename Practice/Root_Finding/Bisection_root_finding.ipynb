```python
import pandas as pd

def f(expr,val):
    return eval(expr,{'x':val})

def bisection(expr,low,high,et):
    fl=f(expr,low)
    fh=f(expr,high)
    if(fl*fh>0):
      raise Exception('No solution, wrong guess!')
    
    mid=(low+high)/2
    data=[]
    data.append([round(low,5),round(high,5),round(mid,5),f(expr,mid),  " -   "])
    indigator=f(expr,low)*f(expr,mid)
    if indigator<=0:
      high=mid
    else:
      low=mid
    error=100
    flag=error>et

    while(flag):
        it=1
        cur_mid=(high+low)/2
        indicator=f(expr,low)*f(expr,cur_mid)

        if indicator<=0:
          high=cur_mid
        else:
          low=cur_mid
        
        if f(expr,cur_mid)==0:
          error=0
        else:
          error=abs((cur_mid-mid)/mid)*100.0
        it=it+1
        data.append([round(low,5),round(high,5),round(mid,5),f(expr,mid), round(error,5)])
        mid=cur_mid
        flag=error>et
    df=pd.DataFrame(data,columns=['Low','High','Mid','F(Xm)','Error(%)'])
    display(df)
    print('Root is : %.6f'%(mid))


# expr='2*x**3+3*x-4'
# low=0
# high=10
# et=0.05

expr=input('Enter the exprrssion in python format:')
low=float(input('Enter the lower guess:'))
high=float(input('Enter the upper guess:'))
et=float(input('Enter the error tolerance:'))

bisection(expr,low,high,et)
```
