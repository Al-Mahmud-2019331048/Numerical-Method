```python
import pandas as pd

def f(expr,val):
    return eval(expr,{"x":val})

def false_position(expr,low,high,et):
    fl=f(expr,low)
    fh=f(expr,high)

    if fl*fh>0:
        raise Exception('No solution, Wrong guess!')
    if fl==fh:
        raise Exception('Divide by zero error!')
    
    mid=(high*fl-low*fh)/(fl-fh)
    data=[]
    data.append([round(low,5),round(high,5),round(mid,5), f(expr,mid), "  -  "])
    indicator=f(expr,low)*f(expr,mid)
    if indicator<=0:
        high=mid
    else:
        low=mid

    error=100
    flag=error>et

    while(flag):
        it=1
        cur_mid=(high*f(expr,low)-low*f(expr,high))/(f(expr,low)-f(expr,high))
        indicator=f(expr,low)*f(expr,cur_mid)

        if indicator<=0:
            high=cur_mid
        else:
            low=cur_mid
        if f(expr,cur_mid)==0:
            error=0
        else:
            error=abs((cur_mid-mid)/cur_mid)*100
        mid=cur_mid
        data.append([round(low,5),round(high,5),round(mid,5),f(expr,mid),round(error,5)])
        it=it+1
        flag=error>et
    df=pd.DataFrame(data,columns=['Low','High','Mid','F(xm)','Error(%)'])
    display(df)
    print('Root is : %.6f'%(mid))
        


# expr='2*x**3+3*x-4'
expr='x**3-.165*x**2+0.0003993'
low=0
high=0.11
et=.02
false_position(expr,low,high,et)
```
