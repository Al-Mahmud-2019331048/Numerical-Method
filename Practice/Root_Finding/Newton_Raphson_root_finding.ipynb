```python
import pandas as pd

def f(expr,val):
    return eval(expr,{'x':val})

def Newton_Raphson(expr,root,et):
    error=100
    flag=error>et
    data=[]
    data.append([round(root,5),"  -  "])
    while(flag):
        it=1
        cur_root=root-f(expr,root)/f(derivative,root)
        error=abs((cur_root-root)/cur_root)*100
        root=cur_root
        data.append([round(root,5),round(error,5)])
        it=it+1
        flag=error>et

    df=pd.DataFrame(data,columns=['Root','Error(%)'])
    display(df)
    print('Root is : %0.6f'%(root))


expr='x**3-.165*x**2+0.0003993'
derivative='3*x**2-2*0.165*x'
root=0.05
et=0.05

Newton_Raphson(expr,root,et)

```
