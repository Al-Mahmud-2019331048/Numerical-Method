```python
import numpy as np

n=int(input('Enter number of unknowns:'))
# input vector
a=np.zeros((n,n+1))
# solution vector
x=np.zeros(n)

# reading augmented matrix coefficinets
print('Enter augmented matrix coefficients:')
for i in range(n):
  for j in range(n+1):
    a[i][j]= float(input('a['+str(i)+']['+str(j)+']='))

# Applying gaussian elimination

for i in range(n):
    if a[i][i]==0:
        raise Exception('Divide by zero error!')
    for j in range(i+1,n):
        ratio=a[j][i]/a[i][i]
        for k in range(n+1):
            a[j][k]=a[j][k]-ratio*a[i][k]

# Back Substitution
x[n-1]=a[n-1][n]/a[n-1][n-1]

for i in range(n-2,-1,-1):
    x[i]=a[i][n]
    for j in range(i+1,n):
        x[i]=x[i]-a[i][j]*x[j]
    x[i]=x[i]/a[i][i]

print('Requred solution is:')
for i in range(n):
    print('X%d : %.3f'%(i,x[i]),end='\t')


```
