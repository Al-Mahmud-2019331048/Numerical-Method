```python
import numpy as np

n=int(input('Enter number of unknowns:'))
# input vector
a=np.zeros((n,n+1))
# solution vector
x=np.zeros(n)


print('Enter augmented coefficinets:')
for i in range(n):
  for j in range(n+1):
    a[i][j]=float(input('a['+str(i)+']['+str(j)+']='))

# Applying Gaussian Elimination by Pivot Method
for i in range(n-1):

    #Find the row with maximum element in current column
    maxElement=abs(a[i][i])
    maxRow=0
    for j in range(i+1,n):
        if maxElement < abs(a[j][i]):
            maxRow=j
            maxElement=abs(a[j][i])
    # Swap the row j with row i
    for j in range(n+1):

        tmp=a[i][j]
        a[i][j]=a[maxRow][j]
        a[maxRow][j]=tmp

    if a[i][i]==0:
        raise Exception("Divide by zero error!")
    
    # Obtain the ratio of matrix[j][i] and matrix[i][i]
    for j in range(i+1,n):
        ratio=a[j][i]/a[i][i]
        # Multiply equation i by ratio and subtract from equation j as well as update the equation j
        for k in range(n+1):
            a[j][k]=a[j][k]-ratio*a[i][k]
        
# Back substitution
x[n-1]=a[n-1][n]/a[n-1][n-1]

# Solution for Xn - 1 to X1

for i in range(n-2,-1,-1):
    x[i]=a[i][n]
    for j in range(i+1,n):
        x[i]=x[i]-a[i][j]*x[j]
    x[i]=x[i]/a[i][i]


print('Requred solution is:')
for i in range(n):
    print('X%d : %.3f'%(i,x[i]),end='\t')
```
